```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
Dataset1 = pd.read_csv('/content/drive/MyDrive/MyPersonalDataset/diabetes.csv')
Dataset1.head()
```

```python
# missing values
Dataset1.isnull().sum()
```

```python
# separating the input and output columns
X = Dataset1.drop(columns = ['Outcome'])
X.head()
```

```python
# import the library
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 5)
pca.fit(X)
X_new = pca.transform(X)
```
