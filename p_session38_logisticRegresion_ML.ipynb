```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
Data = pd.read_csv("/content/drive/MyDrive/MyPersonalDataset/placement.csv")
Data.head()
```

```python
Data.info()
```

```python
Data.shape
```

```python
# missing values
Data.isnull().sum()
```

```python
# plot some figures
plt.scatter(Data["cgpa"], Data["placement_exam_marks"])
```

```python
# split out dataset
X = Data[['cgpa','placement_exam_marks']]
```

```python
X.head()
```

```python
# output data
y = Data[['placed']]
y.head()
```

```python
# train test splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 42) # train_size = 0.75 <=> test_size = 0.25
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# model building
from sklearn.linear_model import LogisticRegression

# creating the object of LogisticRegression
model1 = LogisticRegression()

# training phase
model1.fit(X_train, y_train)
```

```python
# testing phase = performance evaluation
y_pred = model1.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
