# Bar Plots
- used for categorical-numerical analysis

```python
pc = df.pclass.value_counts()
pc
```

```python
plt.bar(x = pc.index, height = pc.values, width = 0.8)
plt.xlabel('Class of people', size = 15)
plt.ylabel('Total count',size = 15)
plt.xticks(color = 'red')
plt.title("Visualization of pclass column", size = 20, color = 'red')
plt.show()
```

```python
plt.figure(figsize = (20,6))
plt.bar(x = pc.index, height = pc.values, width = 0.8, color = ['red','green','black'])
plt.xlabel('Class of people', size = 15)
plt.ylabel('Total count',size = 15)
plt.xticks(color = 'red', rotation = 45, size = 15)
plt.title("Visualization of pclass column", size = 20, color = 'red')
plt.show()
```

## Horizontal Bar charts

```python
plt.figure(figsize = (20,6))
plt.barh(y = pc.index, width = pc.values, height = 0.8, color = ['red','green','black'])
plt.xlabel('Total count', size = 15)
plt.ylabel('Class of people',size = 15)
plt.xticks(color = 'red', rotation = 45, size = 15)
plt.title("Visualization of pclass column", size = 20, color = 'red')
plt.show()
```

## Stacked Bar Plots

```python
x =  ['A','B','C','D','E','F']
y1 = [10,8,6,7,9,15]
y2 = [11,4,8,9,12,3]

plt.bar(x = x, height = y1, color = '#343bba', label = 'y1', alpha = 0.5)
plt.bar(x = x, height = y2, color = 'red', label = 'y2', alpha = 0.5, bottom = y1)
plt.legend(loc = 0)     # loc = 0 will take the best location for legend
plt.show()
```

```python
x =  ['A','B','C','D','E','F']
y1 = [10,8,6,7,9,15]
y2 = [11,4,8,9,12,3]
y3 = [3,5,7,2,8,9]

plt.bar(x = x, height = y1, color = '#343bba', label = 'y1', alpha = 0.5)
plt.bar(x = x, height = y2, color = 'red', label = 'y2', alpha = 0.5, bottom = y1)
plt.bar(x = x, height = y3, color = 'yellow', label = 'y3', alpha = 0.5, bottom = np.sum([y1,y2],axis = 0))

plt.legend(loc = 0)     
plt.show()
```

# Histograms
- is used to visualize the frequnecy of numeric/categorical data

```python
# taking numeric column(age)

plt.hist(x = df.age, bins = 20)
plt.show()
```

```python
# taking categorical column(pclass)

plt.hist(x = df.pclass, bins = 10)
plt.show()
```

```python
# stacking histograms

y1 = [1,2,3,4,5]
y2 = [1,2,3,4,5]

plt.hist(x = y1, bins = 5)
plt.hist(x = y2, bins = 5, alpha = 0.6, bottom = y1, stacked = True)
plt.show()
```

# Line Charts

```python
age = df['age'].sort_values()

plt.plot(range(len(age)), age)
plt.show()
```

# Scatter Plots
- used to check **distribution of data**
- used for continuous data
- also useful in **detecting outliers**

```python
x = np.arange(1,101)
y = np.random.randint(1,100,100)
```

```python
plt.figure(figsize = (10,7))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.scatter(x,y, s = y, c = y)   # s = size, c = color
plt.title("Scatter Plot", size = 20)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.xticks(size = 15, color = 'red')
plt.savefig('myscatterplot.jpeg', dpi = 100);
```

# Box Plots

```python
y = [20,25,26,21,1,28,22,22,29,27,40,70]

plt.boxplot(y)
plt.show()
```

# Pie Charts
- for visualizing the proportions in the data

```python
x = ['Apple','Banana','Cherry','Strawberry','Kiwi']
y = [200,150,100,180,50]

plt.pie(y , explode = [0.1,0.6,0.2,0.3,0.5], labels = x, autopct = "%.2f", shadow = True, radius = 2)
plt.show()
```

# Subplots
- making multiple plots within a single plot

```python
x = range(1,6)
y1 = np.random.randint(1,10,5)
y2 = np.random.randint(1,10,5)

fig, ax = plt.subplots(1,2)   # creating empty subplots area
```

```python
x = range(1,6)
y1 = np.random.randint(1,10,5)
y2 = np.random.randint(1,10,5)

fig, (ax1,ax2) = plt.subplots(1,2)   # creating empty subplots area

ax1.plot(x,y1)
ax2.bar(x,y2)
plt.show()
```

```python
x = range(1,6)
y1 = np.random.randint(1,10,5)
y2 = np.random.randint(1,10,5)

fig, ax = plt.subplots(1,2, sharey = True)   # creating empty subplots area

ax[0].plot(x,y1)
ax[1].bar(x,y2)
plt.show()
```

```python
x = range(1,6)
y1 = np.random.randint(1,10,5)
y2 = np.random.randint(1,10,5)

fig, ax = plt.subplots(2,2, figsize = (8,7))   # creating empty subplots area

ax[0,0].plot(x,y1)
ax[1,1].bar(x,y2)

# adjusting the subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```
