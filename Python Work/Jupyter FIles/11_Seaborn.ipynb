# Seaborn
- is also a data visualization library which is build on top of matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns     # import seaborn library
```

# Load the dataset from seaborns *load_dataset* module

```python
sns.get_dataset_names()    # it will give you the datasets name
```

```python
df = sns.load_dataset('diamonds')
df
```

# Bar Plots

```python
df.cut.value_counts()
```

```python
cut_price = df.groupby(by = 'cut')['price'].mean()

x = cut_price.index
y = cut_price.values
```

```python
sns.barplot(x=x, y = y)
plt.show()
```

```python
# second method

sns.barplot(x = 'cut', y = 'price',data = df)
plt.show()
```

```python
plt.figure(figsize = (15,7))
sns.barplot(x = 'cut', y = 'price',hue='color',data = df)
plt.xlabel("CUT",size = 20)
plt.show()
```

# Line Plots

```python
x = range(1,11)
y1 = np.random.randint(1,10,10)
y2 = np.random.randint(1,10,10)

sns.lineplot(x = x,y = y1, label = 'y1')
sns.lineplot(x = x,y = y2, label = 'y2')
plt.legend()
plt.show()
```

# Scatter Plot

```python
x = range(1,11)
y1 = np.random.randint(1,10,10)
y2 = np.random.randint(1,10,10)

sns.scatterplot(x=x,y=y1, color = 'blue')
sns.scatterplot(x=x,y=y2, marker = "^", color = 'red')
plt.show()
```

# Distribution Plot

```python
import warnings
warnings.filterwarnings('ignore')
```

# Reg Plot
- fits a best fit line on the dataset

```python
x = list(range(1,11))
y = [4, 3, 2, 5, 4, 5, 3, 4, 7, 3]
```

```python
sns.regplot(x = x,y = y)
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.show()
```
