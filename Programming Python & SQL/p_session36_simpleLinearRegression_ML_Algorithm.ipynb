```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
Dataset1 = pd.read_csv("/content/drive/MyDrive/MyPersonalDataset/data.csv")
```

```python
Dataset1.head()
```

```python
Dataset1.shape
```

```python
# missing values
Dataset1.isnull().sum()
```

```python
# let's separate the input and output data
# input data -> capital X
X = Dataset1[['YearsExperience']]
```

```python
X.head()
```

```python
# output data small y
y = Dataset1[['Salary']]
```

```python
y.head()
```

```python
# now we are going to implement the train-test-split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X.shape
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()  # object of linear regression
```

```python
# now we are going to pass our model into training phase --- fit()
```

```python
model1.fit(X_train,y_train) # training phase
```

```python
# now we will check the accuracy of our model
```

```python
X_test
```

```python
# prediction part
```

```python
y_pred = model1.predict(X_test)  # predicted value
```

```python
y_pred
```

```python
y_test # actual ouput
```

```python
# to check the accuracy score
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
