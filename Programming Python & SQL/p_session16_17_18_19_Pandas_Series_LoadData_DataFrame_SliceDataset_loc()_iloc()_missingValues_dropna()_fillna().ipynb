Creating the pandas series:
1. using list
2. using tuple
3. using array
4. using dictionary

```python
# Creating the pandas series from list
series1 = pd.Series([40, 90, 50, 34, 54, 78])
print(series1)
```

```python
# Creating the pandas series from tuple
series2 = pd.Series((50, 90, 34, 55, 600, 1000))
print(series2)
```

```python
# Creating the pandas series from array
import numpy as np
arr1 = np.array([80, 900, 87, 92, 32])
print(arr1)
print()
series3 = pd.Series(arr1)
print(series3)
```

```python
# Creating the pandas series from dictionary
series4 = pd.Series({"A":5000, "B":4000, "C":1000, "D":5500})
print(series4)
```

**Dealing with missing values**
1. remove the missing value -> dropna()
2. fill the missing value -> fillna()

```python
dataset2 = pd.read_csv("/content/drive/MyDrive/MyPersonalDataset/diabetes_missing_values.csv")
dataset2.head(10)
```

**Check the missing values in the DataFrame**

```python
dataset2.isnull()
```

```python
dataset2.isnull().sum()
```

```python
dataset2.isnull().sum().sum()
```

**Removing the missing values - dropna()**

```python
dataset2.shape
```

```python
dataset3 = dataset2.dropna()
dataset3
```

```python
dataset3.shape
```

**Filling the missing values -> fillna()**

```python
dataset2.isnull().sum()
```

**Filling with global constant = 0**

```python
dataset4 = dataset2.fillna(0)
dataset4
```

```python
dataset2.shape
```

```python
dataset4.head()
```

**Fill the missing values with the help of mean**

```python
dataset5 = dataset2.fillna(dataset2.mean())
dataset5
```

```python
one_column = dataset2[['Glucose']].fillna(dataset2[['Glucose']].mean())
one_column
```

**Fill the missing values with the help of median**

```python
dataset6 = dataset2.fillna(dataset2.median())
dataset6
```

```python
dataset6.isnull().sum()
```
