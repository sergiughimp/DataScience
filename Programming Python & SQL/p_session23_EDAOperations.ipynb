**now we are going to detele the columns which are having more than 20% of missing values**

```python
dataset1.shape
```

```python
# row
dataset1.shape[0]
```

```python
# column
dataset1.shape[1]
```

```python
dataset1.isnull().sum()
```

```python
null_var = (dataset1.isnull().sum() / dataset1.shape[0]) * 100
```

```python
# I want the columns where we have more than 20% of missing values
drop_columns = null_var[null_var > 20].keys
drop_columns
```

```python
# now detele the column
dataset2 = dataset1.drop(columns = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'])
dataset2.shape
```

```python
dataset1.shape
```

```python
plt.figure(figsize = (30, 30))
sns.heatmap(dataset2.isnull())
```

**Now I will remove the rows which is having the missing values**

```python
dataset3 = dataset2.dropna()
dataset3.shape
```

```python
plt.figure(figsize = (30, 30))
sns.heatmap(dataset3.isnull())
plt.show()
```
