```python
# plt.figure(figsize=(4, 3))

plt.xlabel("No. of days")
plt.ylabel("Temperature")
plt.title("Daywise Temperature", fontsize=20)

plt.plot(Days, Temp, color='red') # full name color
plt.show()
```

**Multichart in the same figure**

```python
max_temp = [55, 34, 35, 76, 12, 34, 74]
min_temp = [23, 21, 29, 51, 6, 31, 29]
days = ["Mon", "Tue", "Wed", "Thur", "Fri", "Sat", "Sun"]

plt.xlabel("No. of days")
plt.ylabel("Temperature")
plt.title("Daywise Temperature")

plt.plot(days, max_temp, color='red', linestyle='--', marker = "*")
plt.plot(days, min_temp, color='yellow', linestyle='-', marker="v")

# apply legends in the chart
plt.legend(["days vs min_temp", "days vs max_temp"], fontsize=8)
plt.show()
```

**Bar Chart()**
1. graph that represent the category of data with rectangular bars. Describe the comparision between the discrete values

Tools: Tableau, Power BI, Google Data Studies

```python
courses = ["C++", "Java", "Python", "DotNet", "GoLang", "HTML"]
learners = [700, 1000, 1500, 700, 750, 570]

plt.bar(courses, learners)
```

```python
plt.bar(courses, learners, color="red")
plt.xlabel("Types of Courses")
plt.ylabel("Types of Learners")
plt.title("Learners enrolled on the courses")
plt.show()
```

```python
# horizontal bar chart

plt.barh(courses, learners)
plt.ylabel("Types of Courses")
plt.xlabel("Number of Learners")
plt.title("Learners enrolled on the courses")

plt.show()
```

```python
# horizontal bar chart
plt.barh(courses, learners, color=['red', 'yellow', 'black', 'blue', 'cyan'])
plt.ylabel("Types of Courses")
plt.xlabel("Number of Learners")
plt.title("Learners enrolled on the courses")

plt.show()
```

```python
plt.barh(courses, learners, color=['red', 'yellow', 'black', 'blue', 'cyan'])
plt.ylabel("Types of Courses")
plt.xlabel("Number of Learners")
plt.title("Learners enrolled on the courses")

# to apply grid to bar chart
plt.grid(linewidth = '1.4', linestyle = '-', alpha = 0.1, color = 'red')
plt.show()
```

Scatter chart
1. plot used to represent the relationship between two variables

Pie chart
1. circular static plot
2. the area of the chart is the total percentage of the given data

Seaborn = python vizualization library, build on matplotlib

heatmap = square matrix
  * corelation = most importat factor
  * positive correlation (directly proportional)
  * negative correlation (indirectly proportional)
  * 0 -> no correlation

joint plot = to show the relationship between two numerical qualities

pair plot = relationship between more than 2 numerical variables

displot = shows the distribution of the numerical columns along with kde function

count plot = use the count plot to see the count of the categorical values of the column

```python
import matplotlib.pyplot as plt

list1 = [50, 98, 76, 23, 54, 12, 87]
list2 = [59, 43, 76, 45, 98, 12, 34]
list3 = [89, 45, 87, 34, 56, 71, 31]

plt.scatter(list1, list2)
plt.show()
```

```python
plt.scatter(list1, list2, color='red')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
plt.scatter(list1, list2, color='red', linewidths=3)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
plt.scatter(list1, list2, color='red', linewidths=3, marker="D")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
plt.scatter(list1, list2, color='red', linewidths=3, marker="D", edgecolor="green")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
plt.scatter(list1, list2, color='red', linewidths=3, marker="D", edgecolor="green", s=200)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python

plt.scatter(list1, list2, color='red', linewidths=3, marker="D", edgecolor="green", s=150)
plt.scatter(list1, list3, color='yellow', linewidths=2, marker="*", edgecolor="cyan", s=200)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

**Pie chart**
1. circular statical plot
2. the area of the chart is the total percentage of the given data

```python
company = ["Google", "Amazon", "Facebook", "Netflix", "Atlassian"]
revenue = [23, 18, 35, 76, 14]
c = ['red', 'cyan', 'pink', 'orange', 'blue']
plt.pie(revenue, labels=company, colors=c)
plt.show()
```

```python
# adding the percentage
plt.pie(revenue, labels=company, colors=c, autopct='%0.2f%%')
plt.show()
```

```python
# explode
# adding the percentage
my_values = [0.0, 0.2, 0.0, 0.0, 0.2]
plt.pie(revenue, labels=company, colors=c, autopct='%0.2f%%', explode = my_values)
plt.show()
```

```python
# shadow
plt.pie(revenue, labels=company, colors=c, autopct='%0.2f%%', explode = my_values, shadow=True)
plt.show()
```

```python
# save picture
plt.pie(revenue, labels=company, colors=c, autopct='%0.2f%%', explode = my_values, shadow=True)
plt.savefig("PieChart.png", bbox_inches = 'tight', pad_inches=2, transparent=True)
```

**Seaborn**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dataset1 = pd.read_csv("/content/drive/MyDrive/MyPersonalDataset/tips.csv")
dataset1.head()
```

**heatmap = square matrix**
1. correlation (dependency) most important factor
2. positive correlation (directly proportional)
3. negative correlation (indirectly proportional)
4. 0 -> no correlation

```python
dataset1.corr()
```

```python
sns.heatmap(dataset1.corr(), annot=True)
plt.show()
```

**Joint plot = to show the relationship between two numerical qualities**

```python
sns.jointplot(x='tip', y='total_bill', data=dataset1)
```

```python
sns.jointplot(x='tip', y='total_bill', data=dataset1, kind='hex')
```

**Pair Plot = relationship between more than 2 numerical variables**

```python
sns.pairplot(dataset1)
plt.show()
```

```python
sns.pairplot(dataset1, hue='sex')
plt.show()
```

```python
sns.pairplot(dataset1, hue="day")
plt.show()
```

**Displot = shows the distribution of the numerical columns along with kde function**

```python
sns.distplot(dataset1['total_bill'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(dataset1['total_bill'], kde = False)
plt.show()
```
