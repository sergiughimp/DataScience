# Multiple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

Dataset1 = pd.read_csv("/content/drive/MyDrive/MyPersonalDataset/tips.csv")
Dataset1.head()
```

```python
Dataset1.shape
```

```python
# missing values
Dataset1.isnull().sum()
```

```python
# splitting the dataset into input and output
X = Dataset1[['total_bill', 'size']]
X
```

```python
y = Dataset1[['tip']]
y
```

# Train Test Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 42) # train_size = 0.75 <=> test_size = 0.25
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
# model building
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
```

```python
# training phase
model1.fit(X_train, y_train)
```

```python
# testing phase
y_pred = model1.predict(X_test)
y_pred[:5] # predicted output
```

```python
# actual output
y_test[:5]
```

```python
# performance checking
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test, y_pred))
```

```python
# mean square error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
# root mean square error
import math
math.sqrt(mse)
```
